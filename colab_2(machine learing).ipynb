```python
## 조건  여러개
# 30대 이상 40이대 이하(and function)
# 추가:&
df[(df['age']>=30) & (df['age']<= 40)]
```

```python
##조건 여러개 30대미만 혹은 40대 초과
df[(df['age']<30) | (df['age']>40)]
```

```python
# in을 통한 포함 조건 걸기
df[(df['job'] == 'student') | (df['job']=='manager')]
```

```python
df[df['job'].apply(lambda x: x in['student','manager'])]  # lamda함수: 각 요소 리스트에 포함하는지
```
