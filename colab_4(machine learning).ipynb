```python
import numpy as np
abalone_df['length_bool'] = np.where(abalone_df['length']>abalone_df['length'].median(),
                                     'length_long',
                                     'length_short')

abalone_df
```

```python
# 그룹변수 2개 이상을 선택하여 총계처리
abalone_df.groupby(['sex','length_bool']).mean()
```

```python
# 간결하게 표현
abalone_df.groupby(['sex','length_bool'])['whole_weight'].mean()
```

```python
# apply 함수 활용
# Dataframe타입의 객체에서 호출가능한 apply함수
# 열 기준 집계 axis=0
# 행 기준 집계 axis=1
```

```python
# 행 기준 집계
abalone_df[['diameter','whole_weight']].apply(np.average, axis=1)
```

```python
# 사용자 함수를 위한 집계
import math
def avg_ceil(x,y,z):
  return math.ceil((x+y+z)/3) # ceil 올림 3.7 => 4, 3.1 => 4

abalone_df[['diameter','height','whole_weight']].apply(lambda x: avg_ceil(x[0],x[1],x[2]),axis=1)
# 열방향
```

```python
def f(x):
  sum_ = x[0]+x[1]+x[2]
  if sum_>1:
    return True
  else:
    return False
abalone_df[['diameter','height','whole_weight']].apply(lambda x: f(x),axis=1)
```
