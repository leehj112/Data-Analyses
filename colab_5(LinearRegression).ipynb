```python
import matplotlib.pyplot as plt 
import numpy as np 

x = 2 * np.random.rand(100,1)   # 0~1사이의 균일분포 => shape: 100 * 1 
y = 4 + 3*x + np.random.randn(100,1)  # normal 분포 mu = 0, var = 1, shape: 100 *
plt.scatter(x, y) 
plt.show() 
```

```python
x_b = np.c_[np.ones((100,1)),x] 
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y) # np.linalg
theta_best 
# 최적의 파라미터 
```

```python
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)),x_new]

prediction = x_new_b.dot(theta_best)
prediction 
```

```python
plt.plot(x_new,prediction,"r-")
plt.plot(x,y,'b.')
plt.axis([0,2,0,15])
plt.show() 
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression() 
lin_reg.fit(x,y)
print(lin_reg.intercept_,lin_reg.coef_) 
```

```python
# 경사하강법: 여러 종류의 문제에서 최적의 방법을 찾을 수 있는 매우 일반적인 최적의 알고리즘 
# 기본 매커니즘은 지정한 비용 함수를 최적화하기 위해 파라미터를 반복적으로 수정하는 것 

# 경사하강법 구현(implementation)
import numpy as np

x = 2*np.random.rand(100,1)
x_b = np.c_[np.ones((100,1)),x]
y = 4+3*x_b + np.random.randn(100,1)

learning_rate = 0.001
iterations = 10000
m = x_b.shape[0] # x의 데이터수 

theta = np.random.randn(2,1) 

for iteration in range(iterations):
    gradients = 2/m*x_b.T.dot(x_b.dot(theta)-y)
    theta = theta - (learning_rate*gradients) 
theta 


```
